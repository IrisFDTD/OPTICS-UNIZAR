# Principal axis

$\vec D= \hat \varepsilon \vec E$

En sistema de coordenadas ${X,Y,Z}$ arbitrario el tensor dieléctrico se escribe:

$
\hat \varepsilon=
\begin{bmatrix}
    \varepsilon_{1 1} & \varepsilon_{1  2} & \varepsilon_{1  3} \\
    \varepsilon_{2  1} & \varepsilon_{2  2} & \varepsilon_{2  3} \\
    \varepsilon_{3  1} & \varepsilon_{3  2} & \varepsilon_{3  3} \\
\end{bmatrix}
$



¿Existe un sistema ${X',Y',Z'}$ tal que el tensor sea diagonal, $\hat \varepsilon_d$?

$
\hat \varepsilon_d=
\begin{bmatrix}
    \varepsilon_{1} & 0 & 0 \\
    0 & \varepsilon_{2} & 0 \\
    0 & 0 & \varepsilon_{3} \\
\end{bmatrix}
$

Para ello definimos la transformación lineal $T$ que nos lleva de un sistema de coordenadas al otro:

$\vec E' = T \vec E$

$\vec D' = T \vec D$

Entonces se cumple que :  

$T \vec D=\hat \varepsilon_dT \vec E$

Si existe la transformación inversa $T^{-1}$ $\implies \vec D=T^{-1}\hat \varepsilon_dT \vec E$

De ahí que: $ \hat \varepsilon=T^{-1}\hat \varepsilon_dT $

Y por lo tanto: $ \hat \varepsilon_d=T\hat \varepsilon T^{-1} $


La transformación $T$ se obtiene mediante diagonalización ya que:

$ \hat \varepsilon T^{-1}=T^{-1} \hat \varepsilon_d $

Si llamamos a $P=T^{-1}$ la relación nos dice que:

$P=
\begin{bmatrix}
    \vec U_{1} & \vec U_{2} & \vec U_{3}
\end{bmatrix}$

Así que: $\hat \varepsilon \vec U_i=\varepsilon_i\vec U_i$

Y de ahí: $|\hat \varepsilon-\varepsilon_i \mathbf{I} |=0$



$\vec D= \hat \varepsilon_d \vec E$

Uniáxico
$
\hat \varepsilon_d=
\begin{bmatrix}
    \varepsilon_{o} & 0 & 0 \\
    0 & \varepsilon_{o} & 0 \\
    0 & 0 & \varepsilon_{E} \\
\end{bmatrix}
$
